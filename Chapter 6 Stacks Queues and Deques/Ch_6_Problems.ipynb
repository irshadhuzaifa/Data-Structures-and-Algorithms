### R-6.7 
What values are returned during the following sequence of queue operations,

```python

dequeue(), enqueue(7),enqueue(6), dequeue(), dequeue(),enqueue(4), dequeue(), dequeue()
```

````python
class ArrayQueue:
    """FIFO queue implementation using a Python list as underlying storage."""
    DEFAULT_CAPACITY = 10  # moderate capacity for all new queues

    def __init__(self):
        """Create an empty queue."""
        self._data = [None] * ArrayQueue.DEFAULT_CAPACITY
        self._size = 0
        self._front = 0

    def __len__(self):
        """Return the number of elements in the queue."""
        return self._size

    def is_empty(self):
        """Return True if the queue is empty."""
        return self._size == 0

    def first(self):
        """
        Return (but do not remove) the element at the front of the queue.

        Raise Empty exception if the queue is empty.
        """
        if self.is_empty():
            raise Empty("Queue is empty")
        return self._data[self._front]

    def dequeue(self):
        """
        Remove and return the first element of the queue (i.e., FIFO).

        Raise Empty exception if the queue is empty.
        """
        if self.is_empty():
            raise Empty("Queue is empty")
        answer = self._data[self._front]
        self._data[self._front] = None  # help garbage collection
        self._front = (self._front + 1) % len(self._data)
        self._size -= 1
        return answer

    def enqueue(self, e):
        """Add an element to the back of queue."""
        if self._size == len(self._data):
            self._resize(2 * len(self._data))  # double the array size
        avail = (self._front + self._size) % len(self._data)
        self._data[avail] = e
        self._size += 1

    def _resize(self, cap):  # we assume cap >= len(self)
        """Resize to a new list of capacity >= len(self)."""
        old = self._data  # keep track of existing list
        self._data = [None] * cap  # allocate list with new capacity
        walk = self._front
        for k in range(self._size):  # only consider existing elements
            self._data[k] = old[walk]  # intentionally shift indices
            walk = (1 + walk) % len(old)  # use old size as modulus
        self._front = 0  # front has been realigned



my_queue = ArrayQueue()
print(my_queue.enqueue(5))
print(my_queue.enqueue(3))
print(my_queue.dequeue())
print(my_queue.enqueue(2))
print(my_queue.enqueue(8))
print(my_queue.dequeue())
print(my_queue.dequeue())
print(my_queue.enqueue(9))
print(my_queue.enqueue(1))
print(my_queue.dequeue())
print(my_queue.enqueue(7))
print(my_queue.enqueue(6))
print(my_queue.dequeue())
print(my_queue.dequeue())
print(my_queue.enqueue(4))
print(my_queue.dequeue())
print(my_queue.dequeue())
````

### C-6.26 
Describe how to implement the double-ended queue ADT using two stacks
as instance variables. What are the running times of the methods?

````python
class Empty(Exception):
    """Error attempting to access an element from an empty container."""
    pass

class Full(Exception):
    """Error attempting to add an element to a full container."""
    pass

class ArrayStack:
    """LIFO Stack implementation using a Python list as underlying storage."""

    def __init__(self):
        """Create an empty stack."""
        self._data = []  # nonpublic list instance

    def __len__(self):
        """Return the number of elements in the stack."""
        return len(self._data)

    def is_empty(self):
        """Return True if the stack is empty."""
        return len(self._data) == 0

    def push(self, e):
        """Add element e to the top of the stack."""
        self._data.append(e)  # new item stored at end of list

    def top(self):
        """Return (but do not remove) the element at the top of the stack.

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data[-1]  # the last item in the list

    def pop(self):
        """Remove and return the element from the top of the stack (i.e., LIFO).

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data.pop()

    
    def transfer(self, S, T):
        for num in range(S.__len__()):
            element = S.top()
            T.push(element)
            S.pop()

        return T
    

class DequeUsingTwoStacks:
    
    def __init__(self):
        # Initialize two empty stacks
        self.stack1 = ArrayStack()  # Stack for front operations
        self.stack2 = ArrayStack()  # Stack for back operations

    def insertFront(self, element):
        """
        Insert an element at the front of the deque.
        """
        self.stack1.push(element)  # Push element onto stack1

    def insertLast(self, element):
        """
        Insert an element at the back of the deque.
        """
        self.stack2.push(element)  # Push element onto stack2

    def deleteFront(self):
        """
        Remove and return the element at the front of the deque.
        """
        # If stack1 is empty, transfer elements from stack2 to stack1
        if not self.stack1._data:
            while self.stack2._data:  # Transfer all elements if stack1 is empty
                self.stack1.push(self.stack2.pop())
        
        # If stack1 is still empty, raise an error
        if not self.stack1._data:
            raise IndexError("Deque is empty")
        
        return self.stack1.pop()  # Pop the top element from stack1

    def deleteLast(self):
        """
        Remove and return the element at the back of the deque.
        """
        # If stack2 is empty, transfer elements from stack1 to stack2
        if not self.stack2._data:
            while self.stack1._data:  # Transfer all elements if stack2 is empty
                self.stack2.push(self.stack1.pop())
        
        # If stack2 is still empty, raise an error
        if not self.stack2._data:
            raise IndexError("Deque is empty")
        
        return self.stack2.pop()  # Pop the top element from stack2


# Example Usage
deque = DequeUsingTwoStacks()
deque.insertFront(1)
deque.insertLast(2)
print(deque.deleteFront())  # Output: 1
print(deque.deleteLast())   # Output: 2
````

By using two stacks, we can efficiently implement a deque with average `O(1)` running time for all operations, although individual operations might occasionally take `O(n)` time due to element transfers. This approach leverages the inherent properties of stacks to manage elements and simulate deque behavior.

````python
class Empty(Exception):
    """Error attempting to access an element from an empty container."""
    pass

class Full(Exception):
    """Error attempting to add an element to a full container."""
    pass

class ArrayQueue:
    """FIFO queue implementation using a Python list as underlying storage."""
    DEFAULT_CAPACITY = 10  # moderate capacity for all new queues

    def __init__(self):
        """Create an empty queue."""
        self._data = [None] * ArrayQueue.DEFAULT_CAPACITY
        self._size = 0
        self._front = 0

    def __len__(self):
        """Return the number of elements in the queue."""
        return self._size

    def is_empty(self):
        """Return True if the queue is empty."""
        return self._size == 0

    def first(self):
        """
        Return (but do not remove) the element at the front of the queue.

        Raise Empty exception if the queue is empty.
        """
        if self.is_empty():
            raise Empty("Queue is empty")
        return self._data[self._front]

    def dequeue(self):
        """
        Remove and return the first element of the queue (i.e., FIFO).

        Raise Empty exception if the queue is empty.
        """
        if self.is_empty():
            raise Empty("Queue is empty")
        answer = self._data[self._front]
        self._data[self._front] = None  # help garbage collection
        self._front = (self._front + 1) % len(self._data)
        self._size -= 1
        return answer

    def enqueue(self, e):
        """Add an element to the back of queue."""
        if self._size == len(self._data):
            self._resize(2 * len(self._data))  # double the array size
        avail = (self._front + self._size) % len(self._data)
        self._data[avail] = e
        self._size += 1

    def _resize(self, cap):  # we assume cap >= len(self)
        """Resize to a new list of capacity >= len(self)."""
        old = self._data  # keep track of existing list
        self._data = [None] * cap  # allocate list with new capacity
        walk = self._front
        for k in range(self._size):  # only consider existing elements
            self._data[k] = old[walk]  # intentionally shift indices
            walk = (1 + walk) % len(old)  # use old size as modulus
        self._front = 0  # front has been realigned
        
    def rotate(self):
        """
        Rotate the queue by moving the first element to the back.

        This method removes the element at the front of the queue and 
        enqueues it at the back, effectively rotating the queue.
        """
        if self.is_empty():
            raise Empty("Queue is empty")
        answer = self._data[self._front]
        self._data[self._front] = None  # help garbage collection
        self._front = (self._front + 1) % len(self._data)
        avail = (self._front + self._size) % len(self._data)
        self._data[avail] = answer
            
        
# Create an instance of ArrayQueue
queue = ArrayQueue()

# Enqueue some elements
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

# Print the first element of the queue
print("First element:", queue.first()) 

# Rotate the queue
queue.rotate()

# Print the first element of the queue after rotation
print("First element after rotation:", queue.first())
````

### P-6.34 
Implement a program that can input an expression in postfix notation (see
Exercise C-6.22) and output its value.

**C-6.22** 
Postfix notation is an unambiguous way of writing an arithmetic expression
without parentheses. It is defined so that if `“(exp1)op(exp2)”` is a
normal, fully parenthesized expression whose operation is op, the postfix
version of this is `“pexp1 pexp2 op”`, where pexp1 is the postfix version of
`exp1` and `pexp2` is the postfix version of `exp2`. The postfix version of a single
number or variable is just that number or variable. For example, the
postfix version of `“((5+2) ∗ (8−3))/4”` is `“5 2 + 8 3 − ∗ 4 /”`. Describe
a nonrecursive way of evaluating an expression in postfix notation.

````python
class Empty(Exception):
    """Error attempting to access an element from an empty container."""
    pass

class ArrayStack:
    """LIFO Stack implementation using a Python list as underlying storage."""

    def __init__(self):
        """Create an empty stack."""
        self._data = []  # nonpublic list instance

    def __len__(self):
        """Return the number of elements in the stack."""
        return len(self._data)

    def is_empty(self):
        """Return True if the stack is empty."""
        return len(self._data) == 0

    def push(self, e):
        """Add element e to the top of the stack."""
        self._data.append(e)  # new item stored at end of list

    def top(self):
        """Return (but do not remove) the element at the top of the stack.

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data[-1]  # the last item in the list

    def pop(self):
        """Remove and return the element from the top of the stack (i.e., LIFO).

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data.pop()


def evaluate_postfix(postfix_exp):
    exp_stack = ArrayStack()

    try:
        for item in postfix_exp.split():
            if item.isdigit():  # If the item is a digit
                exp_stack.push(int(item))  # Push to the stack as an integer
            else:
                # Ensure there are at least two operands
                if len(exp_stack) < 2:
                    raise ValueError(f"Error: insufficient operands for operator '{item}'")

                # When an operator is encountered, pop the last two numbers from the stack
                b = exp_stack.pop()  # Second operand
                a = exp_stack.pop()  # First operand

                # Perform the corresponding operation
                if item == '+':
                    exp_stack.push(a + b)
                elif item == '-':
                    exp_stack.push(a - b)
                elif item == '*':
                    exp_stack.push(a * b)
                elif item == '/':
                    if b == 0:
                        raise ZeroDivisionError("Error: Division by zero")
                    exp_stack.push(a / b)
                else:
                    raise ValueError(f"Error: unknown operator '{item}'")

        if len(exp_stack) != 1:  # Ensure there is exactly one result in the stack
            raise ValueError("Error: too many operands")

        return exp_stack.pop()

    except Exception as e:
        return str(e)

# Example expressions
postfix_exp_valid = "5 2 + 8 3 - * 4 /"
postfix_exp_invalid_operator = "5 2 + 8 3 - * 4 / @"
postfix_exp_missing_operand = "5 2 + 8 * -"

# Testing
print(f"Valid Expression Result: {evaluate_postfix(postfix_exp_valid)}")
print(evaluate_postfix(postfix_exp_invalid_operator))
print(evaluate_postfix(postfix_exp_missing_operand))
````

### P-6.35 
The introduction of Section 6.1 notes that stacks are often used to provide
“undo” support in applications like a Web browser or text editor. While
support for undo can be implemented with an unbounded stack, many
applications provide only limited support for such an undo history, with a
fixed-capacity stack. When push is invoked with the stack at full capacity,
rather than throwing a Full exception (as described in Exercise C-6.16),
a more typical semantic is to accept the pushed element at the top while
“leaking” the oldest element from the bottom of the stack to make room.
Give an implementation of such a LeakyStack abstraction, using a circular
array with appropriate storage capacity. This class should have a public
interface similar to the bounded-capacity stack in Exercise C-6.16, but
with the desired leaky semantics when full.

````python
class Empty(Exception):
    """Error attempting to access an element from an empty container."""
    pass

class LeakyStack:
    """Leaky stack implementation using a circular array."""

    def __init__(self, capacity):
        """Create an empty stack with a fixed capacity."""
        self._data = [None] * capacity  # fixed-size list
        self._capacity = capacity       # maximum capacity of the stack
        self._size = 0                  # current number of elements
        self._top = -1                  # index of the top element

    def __len__(self):
        """Return the number of elements in the stack."""
        return self._size

    def is_empty(self):
        """Return True if the stack is empty."""
        return self._size == 0

    def is_full(self):
        """Return True if the stack is full."""
        return self._size == self._capacity

    def push(self, e):
        """Add element e to the top of the stack. If the stack is full, leak the oldest element."""
        if self.is_full():
            # Leak the oldest element by incrementing _top in a circular manner
            self._top = (self._top + 1) % self._capacity
        else:
            self._top = (self._top + 1) % self._capacity
            self._size += 1
        self._data[self._top] = e  # store new element at the top index

    def top(self):
        """Return (but do not remove) the element at the top of the stack.

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        return self._data[self._top]

    def pop(self):
        """Remove and return the element from the top of the stack.

        Raise Empty exception if the stack is empty.
        """
        if self.is_empty():
            raise Empty("Stack is empty")
        result = self._data[self._top]
        self._data[self._top] = None  # clear the reference
        self._top = (self._top - 1 + self._capacity) % self._capacity  # move _top back in a circular manner
        self._size -= 1
        return result

# Example usage:
stack = LeakyStack(3)
stack.push(1)
stack.push(2)
stack.push(3)
print(stack._data)  # Output: [1, 2, 3]
stack.push(4)
print(stack._data)  # Output: [4, 2, 3] - 1 is leaked
stack.push(5)
print(stack._data)  # Output: [4, 5, 3] - 2 is leaked
print(stack.pop())  # Output: 5
print(stack._data)  # Output: [4, None, 3]
````
